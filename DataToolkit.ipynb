# Theory Questions

Q1 What is NumPy, and why is it widely used in Python?
- NumPy is a Python library for numerical computing. It provides support for arrays and matrices, along with mathematical functions to operate on them. It is widely used because of its speed, efficiency, and ability to handle large datasets.

Q2 How does broadcasting work in NumPy?
- Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding their dimensions.

Q3 What is a Pandas DataFrame?
- A Pandas DataFrame is a two-dimensional, size-mutable, and heterogeneous data structure with labeled axes (rows and columns).

Q4 Explain the use of the groupby() method in Pandas.
- The groupby() method groups data based on specified keys and performs aggregate operations like mean, sum, or count.

Q5 Why is Seaborn preferred for statistical visualizations?
- Seaborn provides an intuitive interface and built-in themes for creating attractive and informative statistical plots.

Q6 What are the differences between NumPy arrays and Python lists?
- NumPy arrays are faster, more memory-efficient, and support vectorized operations, unlike Python lists.

Q7 What is a heatmap, and when should it be used?
- A heatmap is a graphical representation of data where values are depicted by color intensity. It is used to visualize correlations or density.

Q8 What does the term “vectorized operation” mean in NumPy?
- Vectorized operations allow element-wise operations on arrays without explicit loops, improving performance.

Q9 How does Matplotlib differ from Plotly?
- Matplotlib is static and simpler, while Plotly supports interactive and dynamic visualizations.

Q10 What is the significance of hierarchical indexing in Pandas?
- Hierarchical indexing allows working with multi-level indices, enabling easier data selection and analysis.

Q11 What is the role of Seaborn’s pairplot() function?
- The pairplot() function creates pairwise scatterplots to explore relationships between variables.

Q12 What is the purpose of the describe() function in Pandas?
- It provides summary statistics of numeric columns, such as mean, median, and standard deviation.

Q13 Why is handling missing data important in Pandas?
- Missing data can skew results or cause errors in analysis. Handling it ensures accuracy and reliability.

Q14 What are the benefits of using Plotly for data visualization?
- Plotly offers interactivity, advanced visuals, and web-based integration.

Q15 How does NumPy handle multidimensional arrays?
- NumPy provides efficient methods to create, manipulate, and perform operations on multidimensional arrays.

Q16 What is the role of Bokeh in data visualization?
- Bokeh is used for creating interactive, web-ready visualizations with minimal effort.

Q17 Explain the difference between apply() and map() in Pandas.
- apply() is used for row/column-wise operations on DataFrames.
- map() applies functions element-wise on Series.

Q18 What are some advanced features of NumPy?
- Advanced features include broadcasting, slicing, indexing, and support for linear algebra.

Q19 How does Pandas simplify time series analysis?
- Pandas provides date-time indexing, resampling, and rolling window operations for time series.

Q20What is the role of a pivot table in Pandas?
- A pivot table summarizes data, enabling aggregation and organization based on keys.

Q21 Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy arrays store data in contiguous memory blocks, optimizing slicing operations.

Q22 What are some common use cases for Seaborn?
- Seaborn is used for correlation heatmaps, pair plots, categorical plots, and regression visualizations.



# Practical Question

```python
#Q1 How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print(row_sum)
```

```python
#Q2 Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data = {'A': [10, 20, 30], 'B': [40, 50, 60]}
df = pd.DataFrame(data)
mean_value = df['A'].mean()
print(mean_value)
```

```python
#Q3 Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y, color='blue')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Q4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()

```

```python
#Q5 Generate a bar plot using Plotly.

import plotly.express as px
data = {'Category': ['A', 'B', 'C'], 'Values': [10, 20, 30]}
fig = px.bar(data, x='Category', y='Values', title="Bar Plot Example")
fig.show()
```

```python
#Q6 Create a DataFrame and add a new column based on an existing column.
import pandas as pd
data = {'Name': ['Alice', 'Bob'], 'Score': [85, 90]}
df = pd.DataFrame(data)
df['Grade'] = ['A' if score >= 90 else 'B' for score in df['Score']]
print(df)
```

```python
#Q7 Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 * arr2
print(result)
```

```python
#Q8 Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Line Plot Example")
plt.show()
```

```python
#Q9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Score': [85, 60, 95]}
df = pd.DataFrame(data)
filtered_df = df[df['Score'] > 80]
print(filtered_df)
```

```python
#Q10 Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt
data = [10, 20, 20, 30, 40, 40, 50]
sns.histplot(data, bins=5, kde=True)
plt.title("Histogram Example")
plt.show()
```

```python
#Q11 Perform matrix multiplication using NumPy.
import numpy as np
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
result = np.dot(matrix1, matrix2)
print(result)  # Output: [[19, 22], [43, 50]]
```
